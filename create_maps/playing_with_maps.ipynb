# From Cl to map

```python
import healpy as hp
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
%matplotlib inline
```

```python
import healpy as hp
import matplotlib.pyplot as plt
import numpy as np

files = ['/global/u2/i/iameric/class/output/test_cl_lensed.dat']
data = []
for data_file in files:
    data = np.loadtxt(data_file)
#1:l     2:TT                     3:EE                     4:TE                     5:BB                     6:phiphi                 7:TPhi                   8:Ephi
names = ['l','TT','EE','BB','TE','phiphi','Tphi','Ephi']

def remove_norm(Cl):
    return np.array([2*np.pi*1/l/(l+1)*Cl[l-1] for l in range(1,1+len(Cl))])

TT = remove_norm(np.insert(data[:,1],0,0))
EE = remove_norm(np.insert(data[:,2],0,0))
BB = remove_norm(np.insert(data[:,3],0,0))
TE = remove_norm(np.insert(data[:,4],0,0))
testmap = hp.synfast((TT,EE,BB,TE), nside = 1024, new = True)

hp.mollview(testmap[0],unit = r'$\mu K$')


#save maps
#hp.write_map('testmap.fits',testmap)
```

```python
hp.mollview(testmap[1])
hp.mollview(testmap[2])
```

```python
plt.plot(np.log10(BB[1:]),label='BB')
plt.plot(np.log10(EE[1:]),label='EE')
plt.plot(np.log10(TT[1:]),label='TT')
plt.xlim(0,100)
plt.legend()
plt.show()
```

# turn the map back into Cl's

```python
#read in the maps
Tmap = hp.read_map('testmap.fits',0)
Qmap = hp.read_map('testmap.fits',1)
Umap = hp.read_map('testmap.fits',2)

#create power spectra from these maps
TT, EE, BB, TE, EB, TB = hp.anafast((Tmap,Qmap, Umap),lmax = 1000)

def normalize_cl(Cl):
    return np.array([1/(2*np.pi)*l*(l+1)*Cl[l-1] for l in range(1,len(Cl)+1)])
```

```python
plt.plot(np.log10(normalize_cl(BB[1:])),label='BB')
plt.plot(np.log10(normalize_cl(EE[1:])),label='EE')
plt.plot(np.log10(normalize_cl(TT[1:])),label='TT')
plt.xlim(0,100)
plt.legend()
plt.show()
```
